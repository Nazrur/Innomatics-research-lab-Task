```python
def smart_parking(logs, capacity):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1

        # Track peak usage
        if current > peak:
            peak = current

    # Check parking status
    if current > capacity:
        status = "Exceeded Capacity"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Parking Status:", status)
    print("Peak Parking Usage:", peak)


# Sample Input
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
capacity = 50

smart_parking(logs, capacity)
```

```python
# Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    occupied = len(booked_seats)

    # Calculate percentage
    percentage = (occupied / total_seats) * 100

    # Determine status
    if percentage == 100:
        status = "Housefull"
    elif percentage >= 70:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", int(percentage), "%")
    print("Show Status:", status)


# Sample Input
booked = [1] * 150
seat_occupancy(200, booked)
```

```python
# Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    # Base time (5 minutes per km)
    time = distance * 5

    # Traffic conditions
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    else:
        time += 5

    # Weather conditions
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Cloud Server Load Classification System
def server_load(cpu_readings):
    # Calculate average CPU
    avg = sum(cpu_readings) / len(cpu_readings)

    # Classify load
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(avg), "%")
    print("Server Status:", status)


# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
# Smart Classroom Resource Usage Monitor
def resource_monitor(resources):
    overused = []

    # Check each resource
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    # Alert condition
    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(resources)
```

```python
# Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    # Check capacity
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
