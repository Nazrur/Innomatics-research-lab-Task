```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find all employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

# Output
print("Top Performers Eligible for Bonus:")
print(", ".join(top_performers), f"(Score: {max_score})")
```
